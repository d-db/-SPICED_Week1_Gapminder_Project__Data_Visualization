## Draw a histogram with pandas

```python
import pandas as pd


df = pd.read_excel("./data/gapminder_lifeexpectancy.xlsx", sheet_name="Data", index_col=0)
```

```python
# Remove decimal values from the column titles

old_titles = df.columns
new_titles = list()

for title in old_titles:
    title = int(title)
    new_titles.append(title)

df.columns = new_titles
```

```python
lifeexp = df[[1950, 1975, 2000, 2015]]
```

```python
import matplotlib.pyplot as plt
```

```python
# Display the four histograms in four different graphs next to each other

fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))


ax0.hist(lifeexp[1950], color='red', alpha=0.4);
ax0.set_xlim([20, 90])
ax0.set_ylim([0, 60])
ax0.text(52, 55, f'Median LE: {int(round(lifeexp[1950].median(), 0))} years', fontsize=10)
ax0.axvline(lifeexp[1950].median(), color='k', linestyle='dashed', linewidth=1)
ax0.set_title('1950', fontsize=20)

ax1.hist(lifeexp[1975], color='blue', alpha=0.4);
ax1.set_xlim([20, 90])
ax1.set_ylim([0, 60])
ax1.text(33, 55, f'Median LE: {int(round(lifeexp[1975].median(), 0))} years', fontsize=10)
ax1.axvline(lifeexp[1975].median(), color='k', linestyle='dashed', linewidth=1)
ax1.set_title('1975', fontsize=20)

ax2.hist(lifeexp[2000], color='green', alpha=0.4);
ax2.set_xlim([20, 90])
ax2.set_ylim([0, 60])
ax2.text(40, 55, f'Median LE: {int(round(lifeexp[2000].median(), 0))} years', fontsize=10)
ax2.axvline(lifeexp[2000].median(), color='k', linestyle='dashed', linewidth=1)
ax2.set_title('2000', fontsize=20)

ax3.hist(lifeexp[2015], color='grey', alpha=0.4);
ax3.set_xlim([20, 90])
ax3.set_ylim([0, 60])
ax3.text(42, 55, f'Median LE: {int(round(lifeexp[2015].median(), 0))} years', fontsize=10)
ax3.axvline(lifeexp[2015].median(), color='k', linestyle='dashed', linewidth=1)
ax3.set_title('2015', fontsize=20)

#Setting x and y labels for the figure object
fig.supxlabel('Life Expectancy', fontsize=20)
fig.supylabel('Number of countries', fontsize=20);
```

```python
#Years 1950 and 2015 in one graph to stress the develpment

fig, ax = plt.subplots(figsize = (10,6))

ax.hist(lifeexp[1950], color='green', label='1950', alpha=0.6)
ax.hist(lifeexp[2015], color='blue', label='2015', alpha=0.6)

ax.axvline(lifeexp[1950].median(), color='k', linestyle='dashed', linewidth=1)
ax.axvline(lifeexp[2015].median(), color='k', linestyle='dashed', linewidth=1)

ax.set_xlim([20, 90])
ax.set_ylim([0, 60])

ax.set_xlabel('Life Expectancy', fontsize=15)
ax.set_ylabel('Number of countries', fontsize=15)

#Add the arrow and text
ax.arrow(x=51, y=45, dx=20, dy=0, width=0.5)
ax.text(62, 48, f'Median LE increased from \n{int(round(lifeexp[1950].median(), 0))}\
 to {int(round(lifeexp[2015].median(), 0))} years', fontsize=12, ha="center")

ax.legend(fontsize=15);
```
