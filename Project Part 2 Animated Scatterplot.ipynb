# Step 10

```python
#Add the continent data
cont = pd.read_csv('./data/continents.csv', sep=";")
df = df.merge(cont)
```

```python
df.rename(columns={"country": "Country name",
                   "population": "Population size", 
                   "continent": "Continent",
                   "life_expectancy": "Life expectancy",
                   "fertility_rate": "Fertility rate"
                  }, inplace=True)
```

```python
df
```

# Step 11 and 12

```python
for i in range(1960, 1961):
    dims = (12, 8)
    fig, ax = plt.subplots(figsize=dims)
    df_subset = df.loc[df['year'] == i]
    
    sns.set_style("whitegrid")
    ax = sns.scatterplot(x='Life expectancy', y='Fertility rate',
                data=df_subset, size='Population size', hue='Continent')
    sns.move_legend(ax, "upper right", fontsize=13)
    
    props = dict(boxstyle="round", facecolor='blue', alpha=0.1)
    ax.text(0.05, 0.07, str(i), transform=ax.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)
    plt.title("Correlation between fertility rate and life expectancy (1960 - 2015)", fontsize=20, fontweight="bold")
    plt.xlabel("Life expectancy", fontsize=15)
    plt.ylabel("Fertility rate", fontsize=15)

    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
        
    plt.axis((20,90,0,9))
    
    path = "./data/plots/lifeexp" + str(i) + ".png"
    plt.savefig(path)
    plt.close()
```

## Extra: Working on the way population is depicted in the legend

```python
df_copy = df.copy()
```

```python
df_copy
```

```python
row = df_copy["Population size"].tolist()
```

```python
new_row = list()

for i in row:
    if i <= 100_000_000:
        new_row.append(0.1)
    elif i <= 500_000_000:
        new_row.append(0.5)
    else:
        new_row.append(1.5)
```

```python
df_copy["Population in Billion (upto)"] = new_row
```

```python
df_copy["Population in Billion (upto)"].value_counts()
```

```python
for i in range(1960, 2016):
    dims = (12, 8)
    fig, ax = plt.subplots(figsize=dims)
    df_subset = df_copy.loc[df['year'] == i]
    
    sns.set_style("whitegrid")
    ax = sns.scatterplot(x='Life expectancy', y='Fertility rate',
                data=df_subset, size="Population in Billion (upto)", hue='Continent')
    sns.move_legend(ax, "upper right", fontsize=13)
    
    props = dict(boxstyle="round", facecolor='blue', alpha=0.1)
    ax.text(0.05, 0.07, str(i), transform=ax.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)
    plt.title("Correlation between fertility rate and life expectancy (1960 - 2015)", fontsize=20, fontweight="bold")
    plt.xlabel("Life expectancy", fontsize=15)
    plt.ylabel("Fertility rate", fontsize=15)

    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
        
    plt.axis((20,90,0,9))
    
    path = "./data/plots/new/lifeexp_new" + str(i) + ".png"
    plt.savefig(path)
    plt.close()
```
